```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
cat_features = data[cat_cols]
unprocessed_cat_features = cat_features.select_dtypes(include=[object]).columns.tolist()
unprocessed_cat_features
```

```python
encoder = OneHotEncoder(handle_unknown='ignore')

one_hot_encoded = encoder.fit_transform(data[unprocessed_cat_features]).toarray()
one_hot_data = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(unprocessed_cat_features), index=data.index)
one_hot_data
```

```python
one_hot_encoded_test = encoder.transform(data_test[unprocessed_cat_features]).toarray()
one_hot_data_test = pd.DataFrame(one_hot_encoded_test, columns=encoder.get_feature_names_out(unprocessed_cat_features), index=data_test.index)
one_hot_data_test
```

```python
data_encoded = pd.concat([data, one_hot_data], axis=1)
data_encoded = data_encoded.drop(unprocessed_cat_features, axis=1)
data_encoded
```

```python
data_encoded_test = pd.concat([data_test, one_hot_data_test], axis=1)
data_encoded_test = data_encoded_test.drop(unprocessed_cat_features, axis=1)
data_encoded_test
```

```python
X = data_encoded.drop(columns=['Churn']).values
y = data_encoded['Churn'].values
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, shuffle=True)
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_valid_scaled = scaler.transform(X_valid)
```

```python
# X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.3, shuffle=True)
```

```python
#   C  
Cs = [100, 10, 1, 0.1, 0.01, 0.001]

#      C
model = LogisticRegressionCV(
    Cs=Cs,  #   C  
    scoring='roc_auc',  #   
    random_state=42,  #  
    refit=True  #        C
)

#     
model.fit(X_train_scaled, y_train)

#    
y_pred_proba = model.predict_proba(X_valid_scaled)[:, 1]

#     ROC-AUC
roc_auc = roc_auc_score(y_valid, y_pred_proba)
print(f'ROC-AUC   : {roc_auc:.4f}')

#    C
best_C = model.C_
print(f'   C: {best_C[0]}')
```

```python
import catboost # : https://catboost.ai/docs
```

```python
X_catboost = data.drop(columns=['Churn'])
y_catboost = data['Churn']
```

```python
X_train_c, X_valid_c, y_train_c, y_valid_c = train_test_split(X_catboost, y_catboost, test_size=0.2, shuffle=True)
```

```python
boosting_model = catboost.CatBoostClassifier(n_estimators=200,
                                             cat_features=cat_cols) #    

boosting_model.fit(X_train_c, y_train_c)

y_train_predicted = boosting_model.predict_proba(X_train_c)[:, 1]
y_test_predicted = boosting_model.predict_proba(X_valid_c)[:, 1]
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
train_auc = roc_auc_score(y_train_c, y_train_predicted)
test_auc = roc_auc_score(y_valid_c, y_test_predicted)

plt.figure(figsize=(10,7))
plt.plot(*roc_curve(y_train, y_train_predicted)[:2], label='train AUC={:.4f}'.format(train_auc))
plt.plot(*roc_curve(y_valid_c, y_test_predicted)[:2], label='test AUC={:.4f}'.format(test_auc))
legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
legend_box.set_facecolor("white")
legend_box.set_edgecolor("black")
plt.plot(np.linspace(0,1,100), np.linspace(0,1,100))
plt.show()
```

```python
n_estimators_list = [50, 100, 200, 300]
learning_rate_list = [0.01, 0.05, 0.1, 0.2]

best_auc = 0
best_params = {}
```

```python
for n_estimators in n_estimators_list:
    for learning_rate in learning_rate_list:
        #     
        model = catboost.CatBoostClassifier(
            n_estimators=n_estimators,
            learning_rate=learning_rate,
            cat_features=cat_cols,
            verbose=0 #   
        )

        #  
        model.fit(X_train_c, y_train_c)

        #    
        y_valid_pred = model.predict_proba(X_valid_c)[:, 1]

        #  ROC-AUC
        roc_auc = roc_auc_score(y_valid_c, y_valid_pred)

        #   
        if roc_auc > best_auc:
            best_auc = roc_auc
            best_params = {'n_estimators': n_estimators, 'learning_rate': learning_rate}

        print(f"n_estimators: {n_estimators}, learning_rate: {learning_rate}, ROC-AUC: {roc_auc:.4f}")

#   
print(f"\nBest ROC-AUC: {best_auc:.4f}")
print(f"Best parameters: {best_params}")
```
