Пункт 2:

```python
n_estimators_list = [50, 100, 200, 300]
learning_rate_list = [0.01, 0.05, 0.1, 0.2]

best_auc = 0
best_params = {}
```

```python
for n_estimators in n_estimators_list:
    for learning_rate in learning_rate_list:
        # Создаем модель с текущими параметрами
        model = catboost.CatBoostClassifier(
            n_estimators=n_estimators,
            learning_rate=learning_rate,
            cat_features=cat_cols,
            verbose=0 # Отключаем вывод логов
        )

        # Обучаем модель
        model.fit(X_train_c, y_train_c)

        # Предсказываем на валидационной выборке
        y_valid_pred = model.predict_proba(X_valid_c)[:, 1]

        # Оцениваем ROC-AUC
        roc_auc = roc_auc_score(y_valid_c, y_valid_pred)

        # Сохраняем лучшие параметры
        if roc_auc > best_auc:
            best_auc = roc_auc
            best_params = {'n_estimators': n_estimators, 'learning_rate': learning_rate}

        print(f"n_estimators: {n_estimators}, learning_rate: {learning_rate}, ROC-AUC: {roc_auc:.4f}")

# Вывод лучших параметров
print(f"\nBest ROC-AUC: {best_auc:.4f}")
print(f"Best parameters: {best_params}")
```
